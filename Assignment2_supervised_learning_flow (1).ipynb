```python
#EDA:
#importing libaries: 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

#creating canvas for 2 visualisations.
plt.figure(figsize=(14, 6))
#1) Understanding how the target variable - MedHouseVal, is distributed, and if there is an outliers -By Histogram and Boxplot
#Histogram:
plt.subplot(1, 2, 1) #divide the canvas and embed the first visualistion on the first column. 
sns.histplot(train_table["MedHouseVal"], bins=30, kde=True, color="blue") #making the first visualistion - Histogram
#giving titles for the histogram and the axis
plt.title("Distribution of Median House Value")
plt.xlabel("Median House Value (in $100,000s)")
plt.ylabel("Number of Samples")
#Boxplot
plt.subplot(1, 2, 2) #divide the canvas and embed the second visualistion on the second column. 
sns.boxplot(x=train_table["MedHouseVal"], color="Yellow") #making the second visualistion - Boxplot
#giving titles for the boxplot and the axis X: 
plt.title("Boxplot of Median House Value")
plt.xlabel("Median House Value")
#Conclusion: 
#The goal is to check whether the distribution is normal and whether there are any outliers or an upper cap in the house price values.
# A concentration around relatively low values (1–3) was observed, along with a possible artificial cap near the value 5
```

```python
#2) Checking if there is a linear connection between the median increase and the Median House Value by 
# Calculating the corelation and buliding a scaterplot
# Calculating the corelation:
correlation = train_table["MedInc"].corr(train_table["MedHouseVal"]) 
#Creating a scaterplot: 
#creating canvas for the scaterplot.
plt.figure(figsize=(10, 6))
#creating the scaterplot
sns.scatterplot(x='MedInc', y='MedHouseVal', data=train_table)
#giving titles for the scaterplot and the axis
plt.title(f"Median Income vs. Median House Value\nPearson Correlation: {correlation:.2f}")
plt.xlabel("Median Income (in $10,000s)")
plt.ylabel("Median House Value (in $100,000s)")
plt.grid(True)
plt.show()
#Conclusion: 
#A relatively strong positive correlation was found (approximately 0.69)
#indicating that median income is an important feature.
#However, the relationship appears to weaken at higher income levels, likely due to a ceiling effect in house prices.
```

```python
#3) Checking the corelation between all the Variables, the goal is to see which Var have high corelation bewteen each other
correlation_matrix = train_table.corr() #creating table of corelation between all variables
plt.figure(figsize=(10, 8)) #determine the size of the table
#drwing the heatmap: 
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
#adding titles
plt.title("Correlation Heatmap Between Features")
plt.show()
#Conclusions: 
#The heatmap shows that MedInc has a strong positive correlation with MedHouseVal (~0.69), indicating it is an important feature.
#Additionally, AveRooms and AveBedrms have a strong correlation (~0.79),
#which may suggest some redundancy and could justify engineering a new feature such as a room-to-bedroom ratio.

```

```python
#4) Creating Feature Engineering for High-Corelation Var: AveRooms/AveBedRooms
#To capture the spaciousness of the house and reduce redundancy between highly correlated features.
#creating a new column called RoomRatio that Show the ratio between AveRooms and AveBedrms
train_table["RoomRatio"] = train_table["AveRooms"] / train_table["AveBedrms"]
plt.figure(figsize=(10, 6)) #determine the size of the graph
sns.histplot(train_table["RoomRatio"], bins=30, kde=True, color="mediumseagreen") #showing the distribution of the new var - By Histogram
#adding a titles for the axis and the histogram
plt.title("Distribution of Room-to-Bedroom Ratio")
plt.xlabel("RoomRatio (AveRooms / AveBedrms)")
plt.ylabel("Number of Samples")
plt.grid(True)
plt.show()
#Conclusion: 
#The RoomRatio feature may better capture meaningful differences in house layout.
#Its distribution shows a wide range, suggesting it could help the model distinguish between property types.
```

```python
#5) Checking if the new feature have a corelation with the target feature: MedHouseVal - by scatterplot 
# Calculating the corelation and buliding a scaterplot
# Calculating the corelation:
correlation = train_table["RoomRatio"].corr(train_table["MedHouseVal"]) 
#Creating a scaterplot: 
#creating canvas for the scaterplot.
plt.figure(figsize=(10, 6))
#creating the scaterplot
sns.scatterplot(x='RoomRatio', y='MedHouseVal', data=train_table)
#giving titles for the scaterplot and the axis
plt.title(f"RoomRatio vs. Median House Value\nPearson Correlation: {correlation:.2f}")
plt.xlabel("Room Ratio")
plt.ylabel("Median House Value (in $100,000s)")
plt.grid(True)
plt.show()
#Conclusion: 
#While some positive relationship is visible, especially for RoomRatios between 1 and 5, the data becomes noisy at higher values.
#The feature may offer useful predictive power, but additional cleaning or transformations may improve its effectiveness.
```

```python
#Cleaning the data from outliers: 
# Data Cleaning:
# - Removed entries where AveOccup > 20 (extreme outliers)
# - Removed entries where RoomRatio > 10 (suspiciously high ratio)
# - Removed entries with MedHouseVal == 5 due to artificial value cap

#creating a new var for all the condition, line will be stay in the data only if it meet with all the conditions:
conditions = (
    (train_table["AveOccup"] < 20) &
    (train_table["RoomRatio"] < 10) &
    (train_table["MedHouseVal"] < 5)
)
cleaned_table = train_table[conditions].copy() #sorting the data for only the lines which meet the conditions
num_removed = len(train_table) - len(cleaned_table) #calculating how many outliers removed
print(num_removed)
cleaned_table.head() 
```

```python
# Two regression models were selected for evaluation:
# - Linear Regression: a simple, interpretable model that assumes linear relationships and is sensitive to feature scaling.
# - Random Forest Regressor: a robust, non-linear ensemble method that can model complex relationships and is less sensitive to outliers.
#
# Feature Engineering:
# - StandardScaler was applied to normalize features, which is important for models like Linear Regression.
# - PolynomialFeatures (degree 2) were tested to capture non-linear interactions between features.
# - Feature Selection (SelectKBest) was used optionally to explore whether reducing input dimensions improves performance.
#
# These approaches were selected to compare both linear and non-linear models, assess the importance of feature scaling and transformation,
# and improve the model's ability to generalize.
```

```python
#stage 1: 
# Linear Regression with 5-Fold Cross Validation and Grid Search:
# Tested combinations of StandardScaler, PolynomialFeatures (degree 1 and 2), and fit_intercept settings.
# Polynomial degree 2 significantly improved R² performance,
#indicating that capturing non-linear feature interactions helps the model better explain the target variable.

import pandas as pd
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV, KFold
from sklearn.metrics import make_scorer, r2_score


#Preparing the data X and y labels
X= cleaned_table.drop(columns=["MedHouseVal"])
y=cleaned_table["MedHouseVal"]
#pipline Creating 
pipe = Pipeline([
    ("scale", StandardScaler()),    #Normalization               
    ("poly", PolynomialFeatures()), #Creates new features at level 1 or 2             
    ("model", LinearRegression())   #The base model            
])
# grid define
param_grid = {
    "poly__degree": [1, 2],                      
    "model__fit_intercept": [True, False]        
}
# GridSearchCV runs all possible combinations
# Uses 5-Fold Cross Validation
# Measures performance with R² – the metric chosen according to the guidelines
grid = GridSearchCV(pipe, param_grid, cv=KFold(n_splits=5, shuffle=True, random_state=42),
                    scoring=make_scorer(r2_score), return_train_score=True)
grid.fit(X, y) #running the experiment
results = pd.DataFrame(grid.cv_results_) #Converts the experiment results into a table (DataFrame)
# Selects only the important columns
# sorts by average result (highest R²)
columns = ["param_poly__degree", "param_model__fit_intercept", "mean_test_score", "std_test_score"]
summary = results[columns].sort_values(by="mean_test_score", ascending=False)
summary.reset_index(drop=True, inplace=True)
summary

# Results show that using PolynomialFeatures with degree=2 significantly improved R² performance
# compared to a simple linear model (degree=1). The best configuration achieved an R² of ~0.67,
# suggesting that capturing non-linear interactions improves predictive power.
```

```python
#Stage2: 
# Random Forest Regressor was tested with different hyperparameter combinations (n_estimators, max_depth)
# using 5-Fold Cross Validation. The model was chosen for its ability to capture non-linear relationships
# and robustness to outliers. Results were compared using R².
from sklearn.ensemble import RandomForestRegressor
#Preparing the data X and y labels
X= cleaned_table.drop(columns=["MedHouseVal"])
y=cleaned_table["MedHouseVal"]
#pipline Creating 
pipe = Pipeline([
    ("scale", StandardScaler()), #Normalization
    ("model", RandomForestRegressor(random_state=42, n_jobs=-1)) #the model
])
#grid define
param_grid = {
    "model__n_estimators": [50],
    "model__max_depth": [5, 10]
}
grid = GridSearchCV(pipe, param_grid, cv=KFold(n_splits=5, shuffle=True, random_state=42),
                    scoring=make_scorer(r2_score), return_train_score=True)
grid.fit(X, y) #running the experiment
results = pd.DataFrame(grid.cv_results_) #Converts the experiment results into a table (DataFrame)
# Selects only the important columns
# sorts by average result (highest R²)
columns = ["param_model__n_estimators", "param_model__max_depth", "mean_test_score", "std_test_score"]
summary = results[columns].sort_values(by="mean_test_score", ascending=False)
summary.reset_index(drop=True, inplace=True)
summary
#Conclusion: 
# Random Forest achieved the best R² score of ~0.74 with 50 trees and max depth of 10.
# This outperformed the Linear Regression models, confirming that Random Forest captures more complex relationships in the data.
```

```python
#Stage3:
# As a bonus experiment, we combined SelectKBest (k=5) with XGBoostRegressor
# to evaluate the effect of advanced feature selection and a non-linear boosting model.
# Using 5-Fold Cross Validation, the pipeline achieved an average R² of ~0.64,
# demonstrating strong performance with a reduced feature set.

import pandas as pd
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import make_scorer, r2_score
from xgboost import XGBRegressor

#Preparing the data X and y labels
data = pd.read_csv("housing_train.csv").copy()
data["RoomRatio"] = data["AveRooms"] / data["AveBedrms"]
data = data[(data["AveOccup"] < 20) & (data["RoomRatio"] < 10) & (data["MedHouseVal"] < 5)].copy()
X = data.drop(columns=["MedHouseVal"])
y = data["MedHouseVal"] 

pipe = Pipeline([
    ("scale", StandardScaler()),  # Normalization
    ("select", SelectKBest(score_func=f_regression, k=5)),  # Select top 5 features
    ("model", XGBRegressor(objective='reg:squarederror', # XGBoost regression model
                           random_state=42, n_jobs=-1,
                           n_estimators=50, max_depth=3))
])
cv = KFold(n_splits=5, shuffle=True, random_state=42) #define Cross Validation
#Running the model and checking how good it predict the price with r^2 score
scores = cross_val_score(pipe, X, y, cv=cv, scoring=make_scorer(r2_score))
print(f"Average R²: {scores.mean():.3f}")
print(f"Standard Deviation: {scores.std():.3f}")
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
import pandas as pd
```

```python
#reading the data
train_data = pd.read_csv("housing_train.csv").copy()
train_data["RoomRatio"] = train_data["AveRooms"] / train_data["AveBedrms"]
train_data = train_data[(train_data["AveOccup"] < 20) & (train_data["RoomRatio"] < 10) & (train_data["MedHouseVal"] < 5)].copy()
X_train = train_data.drop(columns=["MedHouseVal"])
y_train = train_data["MedHouseVal"]
final_model = Pipeline([
    ("scale", StandardScaler()),
    ("model", RandomForestRegressor(n_estimators=50, max_depth=10, random_state=42, n_jobs=-1))
])
#fitting the model on the train set
final_model.fit(X_train, y_train)
test_data = pd.read_csv("housing_test.csv").copy()
test_data["RoomRatio"] = test_data["AveRooms"] / test_data["AveBedrms"]
#preparing the test set
X_test = test_data.drop(columns=["MedHouseVal"], errors="ignore")
y_test = test_data["MedHouseVal"] if "MedHouseVal" in test_data.columns else None
#prediction on the test set
predictions = final_model.predict(X_test)
#calculation the quality of the model, and showing the resault of the 5 first rows
r2 = r2_score(y_test, predictions) if y_test is not None else None
first_5 = predictions[:5]
r2, first_5
```
